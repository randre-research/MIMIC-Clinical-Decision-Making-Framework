```python
# After Analysis
# One model
    # "D:\Users\orosh\Documents\Research\Datasets\mimic-iv-ext-cdm-1.1-from-iv\logs\analysis\FULL_INFO_PLI_N_ONLYABNORMAL_BIN_BINABNORMAL_VANILLA_PROBS\Llama-3.2-1B-Instruct-exl2-4.0bpw_stella_en_400M_v5_evals.pkl"
    # "D:\Users\orosh\Documents\Research\Datasets\mimic-iv-ext-cdm-1.1-from-iv\logs\analysis\FULL_INFO_PLI_N_ONLYABNORMAL_BIN_BINABNORMAL_VANILLA_PROBS\Llama-3.2-1B-Instruct-exl2-4.0bpw_stella_en_400M_v5_results.pkl"
    # "D:\Users\orosh\Documents\Research\Datasets\mimic-iv-ext-cdm-1.1-from-iv\logs\analysis\FULL_INFO_PLI_N_ONLYABNORMAL_BIN_BINABNORMAL_VANILLA_PROBS\Llama-3.2-1B-Instruct-exl2-4.0bpw_stella_en_400M_v5_scores.pkl"

#Same as above, but for each of these three tables

PATH = r"D:\Users\orosh\Documents\Research\Datasets\mimic-iv-ext-cdm-1.1-from-iv\logs\analysis\FULL_INFO_PLI_N_ONLYABNORMAL_BIN_BINABNORMAL_VANILLA_PROBS\Llama-3.2-1B-Instruct-exl2-4.0bpw_stella_en_400M_v5_evals.pkl"

m_results = pickle.load(open(PATH, 'rb'))

print(m_results.keys()) #dict_keys(['appendicitis', 'cholecystitis', 'diverticulitis', 'pancreatitis'])

#flatten the dictionary on patients (i.e. keep only one patient key and get rid of the rest)
def flatten_dict(d):
    """
    Recursively explores the structure of a dictionary.
    If a level has keys that are integers (patient IDs), it keeps only the first one.
    """
    # First, identify integer keys (patient IDs)
    int_keys = [k for k in d.keys() if isinstance(k, int) or isinstance(k, np.int64)]
    if int_keys:
        # Keep only the first integer key
        first_key = int_keys[0]
        # Delete other integer keys
        for k in int_keys:
            if k != first_key:
                del d[k]

    # Now, safely iterate over the remaining keys
    for key in list(d.keys()):
        value = d[key]
        if isinstance(value, dict):
            flatten_dict(value)
        elif isinstance(value, list):
            for item in value:
                if isinstance(item, dict):
                    flatten_dict(item)


flatten_dict(m_results)

def print_dict_structure(d, indent=0):
    """Recursively prints the structure of a dictionary."""
    """If the key is an int, it is a patient ID, in that case keep only one and get rid of the rest"""
    for key in d:
        print('  ' * indent + f'- {key}')
        if isinstance(d[key], dict):
            print_dict_structure(d[key], indent + 1)
        elif isinstance(d[key], list):
            for index, item in enumerate(d[key]):
                if isinstance(item, dict):
                    print('  ' * (indent + 1) + f'- [List Item {index}]')
                    print_dict_structure(item, indent + 2)

# Print the structure
print_dict_structure(m_results)
```

```python
# Updated patient_to_string function with agent scratchpad support + requery before docs
def patient_to_string_with_scratchpad(patient):
    """
    Enhanced version of patient_to_string that includes agent scratchpad information
    from step_data if available, and prints the RAG requery before the retrieved documents.
    """

    def append_retrieval_section(idx: int):
        """Append a retrieval section # (idx+1), showing the requery (if present) before docs."""
        if not is_rag:
            return

        nonlocal patient_str
        patient_str += f"# --- RETRIEVAL {idx + 1} ---\n"

        # Requery from step_data (if available)
        requery_text = None
        if has_step_data and idx < len(patient.get('step_data', [])):
            step_data_i = patient['step_data'][idx]
            # Prefer top-level 'rag_query'; fallback to inputs_for_prompt if needed
            requery_text = step_data_i.get('rag_query') or step_data_i.get('inputs_for_prompt', {}).get('rag_query')

        if requery_text:
            patient_str += f"Requery: {requery_text}\n\n"

        # Retrieved documents (if available)
        if 'retrieval' in patient and idx < len(patient['retrieval']):
            for chunk in patient['retrieval'][idx]:
                patient_str += f"Document: {chunk.get('document_reference')}\n"
                patient_str += f"Page: {chunk.get('page_number')}\n"
                patient_str += f"Content: {chunk.get('content')}\n\n"

    is_rag = 'retrieval' in patient
    has_step_data = 'step_data' in patient

    patient_str = ""
    patient_str += "# --- INPUT ---\n"
    patient_str += patient['input'] + "\n\n"

    # Retrieval after input corresponds to step_data[0] (i.e., for Step 1)
    append_retrieval_section(0)

    patient_str += "# --- INTERMEDIATE STEPS ---\n"
    for i, (action, log) in enumerate(patient['intermediate_steps']):
        patient_str += f"# --- STEP {i + 1} ---\n"
        patient_str += f"{action}:\n {log}\n\n"

        # Add agent scratchpad if available for this step
        if has_step_data and i < len(patient['step_data']):
            step_data = patient['step_data'][i]
            if 'inputs_for_prompt' in step_data and 'agent_scratchpad' in step_data['inputs_for_prompt']:
                scratchpad = step_data['inputs_for_prompt']['agent_scratchpad']
                patient_str += f"# --- AGENT SCRATCHPAD {i + 1} ---\n"
                patient_str += f"{scratchpad}\n\n"

        # Retrieval after this step corresponds to step_data[i+1]
        append_retrieval_section(i + 1)

    patient_str += "# --- OUTPUT ---\n"
    patient_str += patient['output'] + "\n\n"

    return patient_str
```

```python
#Functionize
def SaveTraj(Diagnosis, Doc_without_rag, Doc_with_rag, Min_Traj_Len, Diff_traj_len):
    #Load the tables
    m_results = read_from_pickle_file(Doc_without_rag)
    m_results = {k: v for d in m_results for k, v in d.items()}

    m_results_rag = read_from_pickle_file(Doc_with_rag)
    m_results_rag = {k: v for d in m_results_rag for k, v in d.items()}

    patients_keys_mentionning_guidelines = []

    #Search for patient that fills the criterias
    #Criteria 1: DIAGNOSIS not found in output of model without RAG but found in output of model with RAG
    #Criteria 2: The patient has at least MIN_TRAJ_LEN intermediate steps in both outputs
    patient_keys = []
    for patient in m_results:
        #Look only for the substring between "Final Diagnosis:" and "Treatment:"
        #If the output does not contain Final Diagnosis or Treatment, skip the patient
        if "Final Diagnosis:" not in m_results[patient]['output'] or "Treatment:" not in m_results[patient]['output']:
            continue
        if "Final Diagnosis:" not in m_results_rag[patient]['output'] or "Treatment:" not in m_results_rag[patient]['output']:
            continue
        sub = m_results[patient]['output'].split("Final Diagnosis:")[1].split("Treatment:")[0]
        sub = sub.strip().replace("\n", " ").lower()
        sub_rag = m_results_rag[patient]['output'].split("Final Diagnosis:")[1].split("Treatment:")[0]
        sub_rag = sub_rag.strip().replace("\n", " ").lower()
        # if DIAGNOSIS in m_results_rag[patient]['output'] and DIAGNOSIS not in m_results[patient]['output']:
        if Diagnosis not in sub and Diagnosis in sub_rag:
            if len(m_results_rag[patient]['intermediate_steps']) >= Min_Traj_Len and len(m_results[patient]['intermediate_steps']) >= Min_Traj_Len:
                if len(m_results_rag[patient]['intermediate_steps']) >= len(m_results[patient]['intermediate_steps']) + Diff_traj_len:
                    patient_keys.append(patient)

                    #if guidelines is mentionned in any of the agent's scratchpads in step_data, add the patient to patients_keys_mentionning_guidelines
                    for step in m_results_rag[patient]['step_data']:
                        if 'inputs_for_prompt' in step and 'agent_scratchpad' in step['inputs_for_prompt']:
                            scratchpad = step['inputs_for_prompt']['agent_scratchpad']
                            if 'guideline' in scratchpad.lower():
                                patients_keys_mentionning_guidelines.append(patient)

                            
    patient_key = patient_keys[0] if patient_keys else None
    print("Patients meeting criteras:", patient_keys)
    print("Patient selected:", patient_key)

    #Iterate over the patient keys and save the patient trajectory
    import os
    output_dir = join(OUTPUT_BASE, "patient_trajectories", MODEL_NAME, Diagnosis)
    os.makedirs(output_dir, exist_ok=True)

    #save patient keys mentionning guidelines in a text file
    with open(join(output_dir, "patients_keys_mentionning_guidelines.txt"), "w") as f:
        for pk in patients_keys_mentionning_guidelines:
            f.write(f"{pk}\n")

    for patient_key in patient_keys:
        output_path = join(output_dir, f"{patient_key}.txt")
        with open(output_path, 'w') as f:
            # f.write(patient_to_string(m_results[patient_key]))
            f.write(patient_to_string_with_scratchpad(m_results[patient_key]))
        output_path_rag = join(output_dir, f"{patient_key}_rag.txt")
        with open(output_path_rag, 'w') as f:
            # f.write(patient_to_string(m_results_rag[patient_key]))
            f.write(patient_to_string_with_scratchpad(m_results_rag[patient_key]))
```

```python
import os
import glob
from pathlib import Path
from typing import Dict, Tuple, Optional, Iterable, Union, Sequence

PATHOLOGIES = ("appendicitis", "cholecystitis", "diverticulitis", "pancreatitis")

def _ensure_iterable_roots(root: Union[str, Path, Sequence[Union[str, Path]]]) -> Iterable[Path]:
    if isinstance(root, (str, Path)):
        return [Path(root)]
    return [Path(r) for r in root]

def _resolve_logs_root(
    base_root: Path,
    model: str,
    path_override: Optional[Dict[str, str]]
) -> Path:
    """
    If base_root already ends with a logs-like dir (logs, logs_* or _seeded), use it directly.
    Otherwise, append override (e.g., logs_small_models) or 'logs'.
    Also routes models containing '_SEED=' into '<logs_root>/_seeded'.
    """
    # # If the provided root already *is* a logs dir, keep it
    # if base_root.name.startswith("logs") or base_root.name == "_seeded":
    #     logs_root = base_root
    # else:
    #     sub = (path_override or {}).get(model, "logs")
    #     logs_root = base_root / sub

    logs_root = base_root

    # If there is an override for this model, use it (replace "logs")
    if path_override and model in path_override:
        sub = path_override[model]
        if sub == "logs" or sub.startswith("logs_"):
            #remove logs at the end
            base_root = base_root.parent if base_root.name == "logs" else base_root
            logs_root = base_root / sub
        else:
            raise ValueError(f"Override for model '{model}' must be 'logs' or start with 'logs_', got '{sub}'")

    # Seeded runs live under .../logs/_seeded/
    if "_SEED=" in model:
        logs_root = logs_root / "_seeded"

    return logs_root

def _build_glob_pattern(logs_root: Path, experiment: str, pathology: str, model: str) -> str:
    """
    Build a glob like:
      <logs_root>/<patho>_ZeroShot_<model>_*_<experiment>/<patho>_ZeroShot_<model>_*_results.pkl
    For CDM_VANILLA (no experiment suffix), folder uses timestamp only.
    """
    file_glob = f"{pathology}_ZeroShot_{model}_*_results.pkl"
    if experiment == "CDM_VANILLA":
        dir_glob = f"{pathology}_ZeroShot_{model}_*[0-9]"  # timestamp-only dir
    else:
        dir_glob = f"{pathology}_ZeroShot_{model}_*_{experiment}"

    return str(logs_root / dir_glob / file_glob)

def _pick_newest_shortest(paths: Iterable[str]) -> Optional[str]:
    """
    From candidate file paths:
      1) choose the SHORTEST path,
      2) if ties, choose the NEWEST by mtime,
      3) if still tied, choose lexicographically smallest for stability.
    """
    candidates = []
    for p in paths:
        try:
            mtime = os.path.getmtime(p)
        except OSError:
            continue
        candidates.append((p, len(p), mtime))
    if not candidates:
        return None
    # shortest first, then newest, then lexicographic
    candidates.sort(key=lambda t: (t[1], -t[2], t[0]))
    return candidates[0][0]


def _find_best_for_model(
    roots: Iterable[Path],
    experiment: str,
    pathology: str,
    model: str,
    path_override: Optional[Dict[str, str]]
) -> Optional[str]:
    all_matches = []
    for base_root in roots:
        logs_root = _resolve_logs_root(base_root, model, path_override)
        pattern = _build_glob_pattern(logs_root, experiment, pathology, model)
        all_matches.extend(glob.glob(pattern))
    return _pick_newest_shortest(all_matches)

def build_files_path_for_pair(
    root: Union[str, Path, Sequence[Union[str, Path]]],
    experiment: str,
    model_no_rag: str,
    model_with_rag: str,
    *,
    path_override: Optional[Dict[str, str]] = None,
    pathologies: Sequence[str] = PATHOLOGIES
) -> Dict[str, Tuple[Optional[str], Optional[str]]]:
    roots = _ensure_iterable_roots(root)
    out: Dict[str, Tuple[Optional[str], Optional[str]]] = {}
    for patho in pathologies:
        no_rag_path = _find_best_for_model(roots, experiment, patho, model_no_rag, path_override)
        rag_path    = _find_best_for_model(roots, experiment, patho, model_with_rag, path_override)
        out[patho] = (no_rag_path, rag_path)
    return out
```
